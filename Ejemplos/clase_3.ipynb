```python
import pandas as pd
import numpy as np
from scipy.fftpack import fft, fftfreq
import matplotlib.pyplot as plt

Datos = pd.read_csv('normal.csv')
titulo='Normal'

# Datos = pd.read_csv('desbalance1.csv')
# titulo='Desbalance 1'

# Datos = pd.read_csv('desbalance2.csv')
# titulo='Desbalance 2'

# Datos = pd.read_csv('desalineamientoh1.csv')
# titulo='Desalineamiento 1'

# Datos = pd.read_csv('desalineamientoh2.csv')
# titulo='Desalineamiento 2'

x=Datos['s3'].to_numpy()
N=len(x)

fs= 51200  # Frecuencia de muestreo/adquisición
dt=1/fs #espaciado temporal
n=30.72 #velocidad de giro en Hz

#Transformada de Fourier
Y = fft(x)[0:int(N/2)]/(N/2)
frq = fftfreq(N,dt)[0:int(N/2)] #vector de frecuencias

plt.figure()
plt.plot(frq,abs(Y)/(2*np.pi*frq),'k')
plt.axvline(x=n, color='r', linestyle='--')
plt.axvline(x=2*n, color='r', linestyle='--')
plt.axvline(x=3*n, color='r', linestyle='--')
plt.xlabel('Frecuencia (Hz)', fontsize=14)
plt.ylabel('Amplitud', fontsize=14)
plt.title(titulo, fontsize=14)
plt.xlim(0,100)
plt.show()
```

# Ejemplo 2

```python
#importar librerías
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert, chirp
from scipy.fftpack import fft, fftfreq

#leer datos
Datos0=sio.loadmat('interno.mat')
x=Datos0['x'][:,0]
t=Datos0['t'][:,0]

#calcular envolvente
env= np.abs(hilbert(x))

#graficar
plt.figure()
plt.plot(t,x,linewidth=1)
plt.plot(t,env,linewidth=1)
plt.xlabel('Tiempo (s)', fontsize=14)
plt.ylabel('Aceleración ' "$(m/s^2)$", fontsize=14)
plt.xlim(0,0.05)
plt.ylim(-4,4)
plt.show()
```

```python
#Transformada de Fourier
L=len(x)
dt=t[2]-t[1]
Fx = fft(x)[0:int(L/2)]/(L/2)
Fe = fft(env)[0:int(L/2)]/(L/2)
frq = fftfreq(L,dt)[0:int(L/2)] 

plt.figure()
plt.plot(frq,abs(Fx))
plt.xlabel('Frecuencia (Hz)', fontsize=14)
plt.ylabel('Aceleración ' "$(m/s^2)$", fontsize=14)
plt.title('Sin envolvente')
plt.xlim(0,200)
plt.ylim(0,0.4)
plt.show()

plt.figure()
plt.plot(frq,abs(Fe))
plt.xlabel('Frecuencia (Hz)', fontsize=14)
plt.ylabel('Aceleración ' "$(m/s^2)$", fontsize=14)
plt.title('Con envolvente')
plt.xlim(0,200)
plt.ylim(0,0.4)
plt.show()
```

```python
n=30
BPFI=5.99

plt.figure()
plt.plot(frq,abs(Fe))
plt.axvline(x=n, color='b', linestyle='-')
plt.axvline(x=2*n, color='b', linestyle='-')
plt.axvline(x=3*n, color='b', linestyle='-')
for i in range(1,40):
    plt.axvline(x=i*BPFI, color='r', linestyle='--')
plt.xlabel('Frecuencia (Hz)', fontsize=14)
plt.ylabel('Aceleración ' "$(m/s^2)$", fontsize=14)
plt.title('Con envolvente')
plt.xlim(0,200)
plt.ylim(0,0.4)
plt.show()
```
